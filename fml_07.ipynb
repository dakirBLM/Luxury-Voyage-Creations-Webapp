### Formulas

* Gini index:
  
$$
G(S) = 1 - \sum_{j=1}^{k} p_{j}^{2}
$$

* where the $p_{j}$  is a fraction of points that belongs to the class j from a set S$

* Gini impurity:
$$
G(S \Rightarrow S_{1},...,S_{r}) = \sum_{i=1}^{r} \frac{S_i}{S} G(S_i)
$$

* Try to split the by attribute by value *3.35 (<= 3.35)*
* Compute the gini index for the left and the right side of the split
    * Left: 0.629
    * Right: 0.285
* Compute the total gini impurity for this split
    * Total: 0.546

    
**How to solve the example?**
$$
Gini(left) = 1 - \left[ \left(\frac{45}{114}\right)^2 + \left(\frac{49}{114}\right)^2 + \left(\frac{20}{114}\right)^2 \right] = 0.629
$$

$$
Gini(right) = 1 - \left[ \left(\frac{5}{36}\right)^2 + \left(\frac{1}{36}\right)^2 + \left(\frac{30}{36}\right)^2 \right] = 0.285
$$

$$
GiniImpurity = \frac{|df_{left}|}{|df|} * Gini(left) + \frac{|df_{right}|}{|df|} * Gini(right) =  \frac{114}{150} * 0.629 + \frac{36}{150} * 0.285 = 0.546
$$

```python
def gini_index(data: pd.DataFrame):
    """Calculate Gini index for the entire dataset"""
    if len(data) == 0:
        return 0

    # Count occurrences of each class
    class_counts = data['species'].value_counts()
    total_count = len(data)

    # Calculate Gini index: 1 - sum(p_j^2)
    gini = 1 - sum((count / total_count) ** 2 for count in class_counts)
    return gini

# Compute Gini index for the whole dataset
print(f"Gini index for entire dataset: {gini_index(df):.3f}")
```

```python
# Split by sepal_width = 3.35 and compute Gini impurity
def calculate_split_gini(df, feature, threshold):
    left_split = df[df[feature] <= threshold]
    right_split = df[df[feature] > threshold]

    if len(left_split) == 0 or len(right_split) == 0:
        return float('inf')

    gini_left = gini_index(left_split)
    gini_right = gini_index(right_split)

    # Calculate weighted Gini impurity
    total_count = len(df)
    gini_impurity = (len(left_split) / total_count) * gini_left + (len(right_split) / total_count) * gini_right

    return gini_impurity, gini_left, gini_right

# Test with sepal_width = 3.35
gini_impurity, gini_left, gini_right = calculate_split_gini(df, 'sepal_width', 3.35)
print(f"Left split Gini: {gini_left:.3f}")
print(f"Right split Gini: {gini_right:.3f}")
print(f"Total Gini impurity: {gini_impurity:.3f}")
```

```python
def find_best_split(df, feature):
    """Find the best split for a given feature"""
    values = np.sort(df[feature].unique())
    boundaries = [(values[i] + values[i+1]) / 2 for i in range(len(values)-1)]

    best_gini = float('inf')
    best_boundary = None

    for boundary in boundaries:
        gini_impurity, _, _ = calculate_split_gini(df, feature, boundary)
        if gini_impurity < best_gini:
            best_gini = gini_impurity
            best_boundary = boundary

    return best_gini, best_boundary

def print_splits(df: pd.DataFrame, column: str) -> None:
    """Print all possible splits for a column and find the best one"""
    values = np.sort(df[column].unique())
    boundaries = [(values[i] + values[i+1]) / 2 for i in range(len(values)-1)]

    best_gini = float('inf')
    best_boundary = None

    for boundary in boundaries:
        gini_impurity, _, _ = calculate_split_gini(df, column, boundary)
        print(f"Gini: {gini_impurity:.3f}, Boundary: {boundary:.2f}")

        if gini_impurity < best_gini:
            best_gini = gini_impurity
            best_boundary = boundary

    print(f"Best -> Gini: {best_gini:.3f}, Boundary: {best_boundary:.2f}")
    return best_gini, best_boundary

# Test for all features
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
best_splits = {}

print("Finding best splits for all features:")
print("=" * 50)
for feature in features:
    print(f"\n{feature}:")
    best_gini, best_boundary = print_splits(df, feature)
    best_splits[feature] = (best_gini, best_boundary)
```

```python
# Find the overall best split
best_feature = None
best_gini = float('inf')
best_threshold = None

for feature, (gini, threshold) in best_splits.items():
    if gini < best_gini:
        best_gini = gini
        best_threshold = threshold
        best_feature = feature

print(f"\nOverall best split:")
print(f"Feature: {best_feature}, Threshold: {best_threshold:.2f}, Gini: {best_gini:.3f}")
```

```python
# Split the data using the best split
df_petal_left = df[df[best_feature] <= best_threshold]
df_petal_right = df[df[best_feature] > best_threshold]

print("Left split class distribution:")
print(df_petal_left.species.value_counts())
print("\nRight split class distribution:")
print(df_petal_right.species.value_counts())
```

```python
# Find best splits for the right branch (which has multiple classes)

print("Finding best splits for right branch:")
print("=" * 50)
for feature in features:
    print(f"\n{feature}:")
    print_splits(df_petal_right, feature)
    print(''.join(['-' for 20]))
```

```python
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split into features and target
X = df.drop('species', axis=1)
y = df['species']

print("Features shape:", X.shape)
print("Target shape:", y.shape)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42, stratify=y)

print(f"Training set size: {len(X_train)}")
print(f"Test set size: {len(X_test)}")

print("\nTest set class distribution:")
print(pd.Series(y_test).value_counts())
```

```python
# Create and train Decision Tree classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.3f}")


# Display the tree structure
print("Decision Tree structure:")
print(export_text(clf, show_weights=True, feature_names=list(X.columns.values)))
```
